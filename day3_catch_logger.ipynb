```python
temperatures = [2.5, 4.0, 6.8, 3.1, 9.2]

for temp in temperatures:
    if temp > 4:
        print(f"⚠️ {temp}°C - FLAG: too warm")
    else:
        print(f"✅ {temp}°C - safe")

'''
Picture a cooler full of seafood crates heading from Lagos Lagoon to a retailer.
An IoT sensor logs temperature every hour.
Instead of a human checking each reading, this loop+conditional pattern auto-flags any batch that broke the safe temperature range
 exactly the kind of automation that makes traceability systems valuable.
 '''
```

```python
# TracceAqua - Cold Chain Inspector

batch_ids = ["BATCH001", "BATCH002", "BATCH003", "BATCH004", "BATCH005"]
temperatures = [2.5, 4.0, 6.8, 3.1, 9.2]

print("===== TracceAqua Cold Chain Report =====")

flagged_count = 0

for i in range(len(batch_ids)):
    batch = batch_ids[i]
    temp = temperatures[i]

    if temp > 4:
        status = "⚠️ FLAGGED"
        flagged_count += 1
    else:
        status = "✅ SAFE"

    print(f"{batch} | {temp}°C | {status}")
    print(f"{flagged_count} batches flagged out of {len(batch_ids)} total")

print("=========================================")
```
