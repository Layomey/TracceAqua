```python
#TracceAqua Batch Price Calculator

batch_id = "BATCH002"
species = "Catfish"
weight_kg = 18.0
price_per_kg = 1500

total_value = weight_kg *  price_per_kg

print(f"Batch {batch_id} ({species}): {weight_kg}kg at N{price_per_kg}/kg = N{total_value} total")
```

```python
crate_capacity_kg = 10
crates_per_batch = weight_kg // crate_capacity_kg
leftover_weight = weight_kg % crate_capacity_kg

print(f"{crates_per_batch} crates gets filled up with {leftover_weight}kg  of {species} left")
```
